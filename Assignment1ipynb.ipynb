**Mark Sheet**

```python
math = float(input("Enter marks 1: "))
urdu = float(input("Enter marks 2: "))

eng= float(input("Enter marks 3: "))
isl = float(input("Enter marks 4: "))
comp = float(input("Enter marks 5: "))

total = math + urdu + eng + isl + comp
percentage = total / 5

print("Total:", total)
print("Percentage:", percentage)

if percentage >= 80:
    grade = "A+"
elif percentage >= 70:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "Fail"

print("Grade:", grade)
```
